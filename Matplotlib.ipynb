```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x)
```

```python
print(y) #sin values
```

```python
print(z) #cos values
```

```python
plt.plot(x, y) #sine wave
plt.show()
```

```python
plt.plot(x, z) #cos wave
plt.show()
```

```python
#adding title x-axis & y-axis labels
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x, z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('cosine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)   
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'g-')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
#for multiple rows
x = np.linspace(-5,5,30)
plt.plot(x, np.sin(x),'g-')
plt.plot(x, np.cos(x),'r--')
plt.show()
```

```python
#Bar plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'French', 'Latin', 'German']
people = [100, 50, 150, 60, 90]
ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#pie chart
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English', 'Spanish', 'French', 'Latin', 'German']
people = [100, 50, 150, 60, 90]
ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

```python
#Scatter plot
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3D scatter plot
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, color='Blue')
plt.show()
```
